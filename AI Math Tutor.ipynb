```python
import sympy as sp

print("AI Math Tutor")
print("Enter a math problem (example: 2+5*3)")

problem = input("Your question: ")

try:
    result = sp.sympify(problem)
    answer = sp.simplify(result)
    print("Step-by-step Solution:")
    print(answer)
except:
    print("Sorry, I cannot solve this problem.")
```
